## Let's look at the top scoring players in 2024-25 season

```python
total_player_points = df_2024.groupby('name')['total_points'].sum()
```

```python
print(total_player_points.sort_values(ascending=False).head(10))
```

```python
playerpm = df_2024.groupby('name').agg({'total_points': 'sum', 'minutes': 'sum', 'position': 'last'}).reset_index()
# filter for players with at least 500 minutes
active_players = playerpm[playerpm['minutes'] >= 500]

# start plot

sns.set_theme(style="whitegrid")
fig, ax = plt.subplots(figsize=(11, 7))

sns.scatterplot(data=playerpm, x='minutes', y='total_points', hue='position', 
                size='total_points', sizes=(20, 200), alpha=0.7)

# add highest point scorers onto the plot
top_scorers = active_players.sort_values(by='total_points', ascending=False).head(5)
for i, row in top_scorers.iterrows():
    ax.text(row['minutes'] + 40, row['total_points'] - 2, row['name'], fontsize=9, fontweight='bold')
```

```python
import numpy as np
import pandas as pd


df = pd.read_csv('data/merged_data.csv')

df = df.sort_values(by=['name', 'GW']).reset_index(drop=True)

# Calculate total points in the next 3 gameweeks (GW1 + GW2 + GW3)
df['target_points_next_3_gws'] = (
    df.groupby('name')['total_points'].shift(-1) +
    df.groupby('name')['total_points'].shift(-2) +
    df.groupby('name')['total_points'].shift(-3)
)

# Calculate total points in the next 5 gameweeks
df['target_points_next_5_gws'] = (
    df.groupby('name')['total_points'].shift(-1) +
    df.groupby('name')['total_points'].shift(-2) +
    df.groupby('name')['total_points'].shift(-3) +
    df.groupby('name')['total_points'].shift(-4) +
    df.groupby('name')['total_points'].shift(-5)
)
```
