```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
playerpm = df.groupby('name').agg({'total_points': 'sum', 'minutes': 'sum', 'position': 'last'}).reset_index()
# filter for players with at least 500 minutes
active_players = playerpm[playerpm['minutes'] >= 500]

# start plot

sns.set_theme(style="whitegrid")
fig, ax = plt.subplots(figsize=(11, 7))

sns.scatterplot(data=playerpm, x='minutes', y='total_points', hue='position', 
                size='total_points', sizes=(20, 200), alpha=0.7)

# add highest point scorers onto the plot
top_scorers = active_players.sort_values(by='total_points', ascending=False).head(5)
for i, row in top_scorers.iterrows():
    ax.text(row['minutes'] + 40, row['total_points'] - 2, row['name'], fontsize=9, fontweight='bold')

```

```python
dfmerged = pd.read_csv('data/merged_data.csv')

playerpm = dfmerged.groupby('name').agg({'total_points': 'sum', 'minutes': 'sum', 'position': 'last'}).reset_index()
# filter for players with at least 500 minutes
active_players = playerpm[playerpm['minutes'] >= 500]

# start plot

sns.set_theme(style="whitegrid")
fig, ax = plt.subplots(figsize=(11, 7))

sns.scatterplot(data=active_players, x='minutes', y='total_points', hue='position', 
                size='total_points', sizes=(20, 200), alpha=0.7, ax = ax)

# add highest point scorers onto the plot
top_scorers = active_players.sort_values(by='total_points', ascending=False).head(5)
for i, row in top_scorers.iterrows():
    ax.text(row['minutes'] + 40, row['total_points'] - 2, row['name'], fontsize=9, fontweight='bold')
```
